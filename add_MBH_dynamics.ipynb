### qin-qout 2d histogram

```python
N = 100

hist_Tr_ej_tot = 0
hist_Tr_tot = 0
hist_no_tot = 0

for i in range(N):
    df_trips_i = df_trips[520*i:520*(i+1)]

    df_Tr_ej_qin = df_trips_i[df_trips_i["merger_flag"]=="Tr-ej"]["qin"].to_numpy()
    df_Tr_ej_qout = df_trips_i[df_trips_i["merger_flag"]=="Tr-ej"]["qout"].to_numpy()

    df_no_qin = df_trips_i[df_trips_i["merger_flag"]=="No"]["qin"].to_numpy()
    df_no_qout = df_trips_i[df_trips_i["merger_flag"]=="No"]["qout"].to_numpy()

    df_Tr_qin = df_trips_i[(df_trips_i["merger_flag"]!="Tr-ej")&(df_trips_i["merger_flag"]!="No")]["qin"].to_numpy()
    df_Tr_qout = df_trips_i[(df_trips_i["merger_flag"]!="Tr-ej")&(df_trips_i["merger_flag"]!="No")]["qout"].to_numpy()

    
    xedges =  np.linspace(-4,3,21) #logspace(aa, 21)
    yedges =  np.linspace(-4,0,21)#logspace(bb, 21)

    aa_Tr_ej, bb_Tr_ej = np.log10(df_Tr_ej_qout),np.log10(df_Tr_ej_qin)
    
    hist_Tr_ej, *_ = sp.stats.binned_statistic_2d(aa_Tr_ej, bb_Tr_ej, None, bins=(xedges, yedges), statistic='count')
    hist_Tr_ej_tot = hist_Tr_ej_tot + hist_Tr_ej


    aa_Tr, bb_Tr = np.log10(df_Tr_qout),np.log10(df_Tr_qin)
    hist_Tr, *_ = sp.stats.binned_statistic_2d(aa_Tr, bb_Tr, None, bins=(xedges, yedges), statistic='count')
    hist_Tr_tot = hist_Tr_tot + hist_Tr


    aa_no, bb_no = np.log10(df_no_qout),np.log10(df_no_qin)
    hist_no, *_ = sp.stats.binned_statistic_2d(aa_no, bb_no, None, bins=(xedges, yedges), statistic='count')
    hist_no_tot = hist_no_tot + hist_no


hist_Tr_ej_tot = hist_Tr_ej_tot/N
hist_Tr_tot = hist_Tr_tot/N
hist_no_tot = hist_no_tot/N
```

```python
fig,axes = plt.subplots(1,3,figsize=[15,4])
xx, yy = np.meshgrid(xedges, yedges)
#axes[0,0] = plt.gca()


#norm = mpl.colors.Normalize(0,round(hist_no_tot.max()))

norm = mpl.colors.Normalize(0,round(hist_Tr_tot.max()))
pcm = axes[0].pcolormesh(xedges, yedges, hist_Tr_tot.T, cmap="Blues", norm=norm)
plt.colorbar(pcm, ax=axes[0], label='number')

norm = mpl.colors.Normalize(0,round(hist_Tr_ej_tot.max()))
pcm = axes[1].pcolormesh(xedges, yedges, hist_Tr_ej_tot.T, cmap="Blues", norm=norm)
plt.colorbar(pcm, ax=axes[1], label='number')


norm = mpl.colors.Normalize(0,round(hist_no_tot.max()))
pcm = axes[2].pcolormesh(xedges, yedges, hist_no_tot.T, cmap="Blues", norm=norm)
plt.colorbar(pcm, ax=axes[2], label='number')


axes[0].set_xlabel("$\log_{10}(q_{out})$")
axes[1].set_xlabel("$\log_{10}(q_{out})$")
axes[2].set_xlabel("$\log_{10}(q_{out})$")

axes[0].set_ylabel("$\log_{10}(q_{in})$")
axes[1].set_ylabel("$\log_{10}(q_{in})$")
axes[2].set_ylabel("$\log_{10}(q_{in})$")

axes[0].set_title("Prompt Mergers")
axes[1].set_title("Mergers after kick")
axes[2].set_title("No mergers")

plt.tight_layout()
plt.savefig("Figures/2d-hist-qin-qout-for-all-cases.pdf")
```

```python
N = 100

qin_mrg_tot = 0
qin_mrg_ejec_tot = 0
qin_no_tot = 0


qout_mrg_tot = 0
qout_mrg_ejec_tot = 0
qout_no_tot = 0

q_min = -4
q_max = 3
q_binsize = 0.6
q_bins = int((q_max-q_min)/q_binsize)


M1_mrg_tot = 0
M1_no_tot = 0

m1_min = 6
m1_max = 10
m1_binsize = 0.25
m1_bins = int((m1_max-m1_min)/m1_binsize)

for i in range(N):
    df_trips_i = df_trips[520*i:520*(i+1)]

    df_no_qin = df_trips_i[df_trips_i["merger_flag"]=="No"]["qin"].to_numpy()
    df_no_qout = df_trips_i[df_trips_i["merger_flag"]=="No"]["qout"].to_numpy()

    df_mrg_ejec_qin = df_trips_i[df_trips_i["merger_flag"]=="Tr-ej"]["qin"].to_numpy()
    df_mrg_ejec_qout = df_trips_i[df_trips_i["merger_flag"]=="Tr-ej"]["qout"].to_numpy()

    df_Tr_qin = df_trips_i[(df_trips_i["merger_flag"]!="Tr-ej")&(df_trips_i["merger_flag"]!="No")]["qin"].to_numpy()
    df_Tr_qout = df_trips_i[(df_trips_i["merger_flag"]!="Tr-ej")&(df_trips_i["merger_flag"]!="No")]["qout"].to_numpy()

    mrg_count,q_bins = np.histogram(np.log10(df_Tr_qin),range=(q_min,q_max),bins = q_bins,density=True)
    qin_mrg_tot = qin_mrg_tot + mrg_count
    mrg_count,q_bins = np.histogram(np.log10(df_Tr_qout),range=(q_min,q_max),bins = q_bins,density=True)
    qout_mrg_tot = qout_mrg_tot + mrg_count

    mrg_count,q_bins = np.histogram(np.log10(df_mrg_ejec_qin),range=(q_min,q_max),bins = q_bins,density=True)
    qin_mrg_ejec_tot = qin_mrg_ejec_tot + mrg_count
    mrg_count,q_bins = np.histogram(np.log10(df_mrg_ejec_qout),range=(q_min,q_max),bins = q_bins,density=True)
    qout_mrg_ejec_tot = qout_mrg_ejec_tot + mrg_count

    mrg_count,q_bins = np.histogram(np.log10(df_no_qin),range=(q_min,q_max),bins = q_bins,density=True)
    qin_no_tot = qin_no_tot + mrg_count
    mrg_count,q_bins = np.histogram(np.log10(df_no_qout),range=(q_min,q_max),bins = q_bins,density=True)
    qout_no_tot = qout_no_tot + mrg_count

    df_mrg_M1 = df_trips_i[df_trips_i["merger_flag"]!="No"]["M1"].to_numpy()
    df_mrg_M2 = df_trips_i[df_trips_i["merger_flag"]!="No"]["M2"].to_numpy()
    df_no_M1 = df_trips_i[df_trips_i["merger_flag"]=="No"]["M1"].to_numpy()
    df_no_M2 = df_trips_i[df_trips_i["merger_flag"]=="No"]["M2"].to_numpy()

    #Mchirp_mrg = (df_mrg_M1*df_mrg_M2)**(3/5)/(df_mrg_M1 + df_mrg_M2)**(1/5)
    Mchirp_mrg = df_mrg_M1 + df_mrg_M2
    Mchirp_no = df_no_M1 + df_no_M2
    #Mchirp_no = (df_no_M1*df_no_M2)**(3/5)/(df_no_M1 + df_no_M2)**(1/5)
    
    #mrg_count,m1_bins = np.histogram(np.log10(df_mrg_M1),range=(m1_min,m1_max),bins = m1_bins,density=True)
    mrg_count,m1_bins = np.histogram(np.log10(Mchirp_mrg),range=(m1_min,m1_max),bins = m1_bins,density=True)
    M1_mrg_tot = M1_mrg_tot + mrg_count

    No_count,No_m1_bins = np.histogram(np.log10(Mchirp_no),range=(m1_min,m1_max),bins = m1_bins,density=True)
    #No_count,No_m1_bins = np.histogram(np.log10(df_no_M1),range=(m1_min,m1_max),bins = m1_bins,density=True)
    M1_no_tot = M1_no_tot + No_count

qin_mrg_tot = qin_mrg_tot/N
qout_mrg_tot = qout_mrg_tot/N

qin_mrg_ejec_tot = qin_mrg_ejec_tot/N
qout_mrg_ejec_tot = qout_mrg_ejec_tot/N

qin_no_tot = qin_no_tot/N
qout_no_tot = qout_no_tot/N

M1_mrg_tot = M1_mrg_tot/N
M1_no_tot = M1_no_tot/N
```

```python
def merger_rate_per_run(df,zmax=4,zbinsize=0.01,lgzbinsize=0.2,lgzmin=-3.0,lgzmax=1.0):

    
    Tr_prompt = df[(df["merger_flag"]=="Tr-12") | (df["merger_flag"]=="Tr-13") | (df["merger_flag"]=="Tr-23")]["z_merger"]
    Tr_prompt_mr,Tr_prompt_lgzbins,Tr_prompt_rate = mr.merger_rate_log_plot(Tr_prompt,zmax,zbinsize,lgzbinsize,lgzmin,lgzmax)


    Tr_ej_z = df[(df["merger_flag"]=="Tr-ej")]["z_merger"]
    Tr_ej_mr,Tr_ej_lgzbins,Tr_ej_rate = mr.merger_rate_log_plot(Tr_ej_z,zmax,zbinsize,lgzbinsize,lgzmin,lgzmax) 

    return Tr_prompt_lgzbins,[Tr_prompt_mr,Tr_prompt_rate,Tr_ej_mr,Tr_ej_rate]
```

```python
def triple_outcomes_merger_rate(df_run,zmax,zbinsize,lgzbinsize,lgzmin,lgzmax):

    net_mr_Tr = []
    net_mr_Tr_ej = []
   

    mr_Tr = []
    mr_Tr_ej = []


    for i in range(Nruns):
        mr_out = merger_rate_per_run(df_run[i],zmax,zbinsize,lgzbinsize,lgzmin,lgzmax)[1:][0]
        net_mr_Tr.append(mr_out[0])
        mr_Tr.append(mr_out[1])
        net_mr_Tr_ej.append(mr_out[2])
        mr_Tr_ej.append(mr_out[3])

    lgzbins = merger_rate_per_run(df_run[i],zmax,zbinsize,lgzbinsize,lgzmin,lgzmax)[0]
    mr_Tr  = np.array(mr_Tr)
    mr_Tr_ej = np.array(mr_Tr_ej)
    net_mr_Tr_ej = np.array(net_mr_Tr_ej)
    net_mr_Tr = np.array(net_mr_Tr)

    return lgzbins,mr_Tr,mr_Tr_ej,net_mr_Tr,net_mr_Tr_ej

def triple_merger_after_ejec_rate(df_run,zmax,zbinsize,lgzbinsize,lgzmin,lgzmax):

    mr_Tr_ej = []
    for i in range(Nruns):
        df = df_run[i]
        Tr_ej_z = df[(df["merger_flag"]=="Tr-ej")]["z_merger"]
        Tr_ej_mr,Tr_ej_lgzbins,Tr_ej_rate = mr.merger_rate_log_plot(Tr_ej_z,zmax,zbinsize,lgzbinsize,lgzmin,lgzmax) 
        mr_Tr_ej.append(Tr_ej_rate)

    return Tr_ej_lgzbins,mr_Tr_ej
```
