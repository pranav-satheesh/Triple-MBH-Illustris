```python
import numpy as np
from scipy.interpolate import RegularGridInterpolator

x = np.array([11,22,33,44,55,66,77])
y = np.array([10**5,10**6,10**7,10**8,10**9,10**10,10**11])
z = np.random.rand(7,7)

x_new = np.array([1,12,23,34,45,56,67])
y_new = np.array([10**1,10**2,10**3,10**4,10**5,10**6,10**7])
```

```python
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits import mplot3d
```

```python
interpolant_function = RegularGridInterpolator((x,y),z,bounds_error=False, fill_value = None)
```

```python
xv, yv = np.meshgrid(x_new,y_new, indexing='ij')
```

```python
result = interpolant_function((xv,yv))
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(xv,yv,result, 50, cmap='binary')
ax.contour3D(xv,yv,z, 50, cmap='viridis')
#surf2 = ax.plot_surface(xv,yv,result,label="OG")
#surf1 = ax.plot_surface(x,y,z,label="extrapolated")
```
